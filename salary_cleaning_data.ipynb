```python
import pandas as pd
```

```python
df = pd.read_csv('D:\Project\ds_salary_prediction\Dataset\glassdoor_jobs.csv')
df
```

**Salary Column**

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)

df = df[df['Salary Estimate'] != '-1']
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.replace('K','000').replace('$',''))
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))

df['average_salary'] = df['Salary Estimate'].apply(lambda x: (int(x.split('-')[0]) + int(x.split('-')[1]))/2)
df['average_salary']
```

```python
df.reset_index()
```

**Company NAme**

```python
df['Company Name'] = df['Company Name'].apply(lambda x: x.split('\n')[0])
df.head()
```

**State**

```python
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
df.job_state.value_counts()
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis = 1)
df['same_state']
```

**No. of years of company**

```python
df['company_age'] = df.Founded.apply(lambda x: x if x <1 else 2021 - x)
df['company_age']
```

**Size**

```python
df['Size'] = df['Size'].apply(lambda x: (x.split('e')[0]).strip())
df['Size']
```

**Type of ownership**

**desc_len**

```python
df['desc_len'] = df['Job Description'].apply(lambda x: len(x))
df['desc_len']
```

**Competitor**

```python
df['Competitors']
df['count_Competitors'] = df['Competitors'].apply(lambda x: len(x.split(',')) if x != '-1' else 0)
df['count_Competitors']
```

**Title**

```python
df['Job Title'].unique()
```

```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'vp' in title.lower():
        return 'vp'
    elif 'engineer' in title.lower():
        return 'data engineer'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    elif 'analyst' in title.lower():
        return 'data analyst'
    elif 'scientist' in title.lower():
        return 'senior/associate scientist'
    elif 'data modeler' in title.lower():
        return 'data modeler'
    elif 'analytics' in title.lower():
        return 'Analytics Consultant'
    else:
        return 'NA'
df['Job Simple'] = df['Job Title'].apply(title_simplifier)
df
```

**Revenue**

```python
df['Revenue'].unique()
```

```python
def revenu_simpler(revenue):
    if 'Less than $1 million (USD)' in revenue:
        return '$0-$1M'
    elif '$1 to $5 million (USD)' in revenue:
        return '$1M-$5M'
    elif '$5 to $10 million (USD)' in revenue:
        return '$5M-$10M'
    elif '$10 to $25 million (USD)' in revenue:
        return '$10M-$25M'
    elif '$25 to $50 million (USD)' in revenue:
        return '$25M-$50M'
    elif '$50 to $100 million (USD)' in revenue:
        return '$50M-$100M'
    elif '$100 to $500 million (USD)' in revenue:
        return '$100M-$500M'
    elif '$500 million to $1 billion (USD)' in revenue:
        return '$500M-$1B'
    elif '$1 to $2 billion (USD)' in revenue:
        return '$1B-$2B'
    elif '$2 to $5 billion (USD)' in revenue:
        return '$2B-$5B'
    elif '$5 to $10 billion (USD)' in revenue:
        return '$5B-$10B'
    elif '$10+ billion (USD)' in revenue:
        return '$10+B'
    else:
        return 'NA'
df['revenue_simpler'] = df['Revenue'].apply(revenu_simpler)
# df[['revenue_simpler','Revenue']]
```
