```python
!pip install wordcloud
import nltk
nltk.download('stopwords')
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# nltk.download('punkt')
words = " ".join(df['Job Description'])

def punctuation_stop(text):
    """remove punctuation and stop words"""
    filtered = []
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    for w in word_tokens:
        if w not in stop_words and w.isalpha():
            filtered.append(w.lower())
    return filtered


words_filtered = punctuation_stop(words)

text = " ".join([ele for ele in words_filtered])

wc= WordCloud(background_color="white", random_state=1,stopwords=STOPWORDS, max_words = 20000, width =1800, height = 500)
wc.generate(text)

plt.figure(figsize=[100,10])
plt.imshow(wc,interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
df_cat = df[['Location', 'Headquarters', 'Size','Type of ownership', 'Industry', 'Sector', 'Revenue', 'job_state','same_state', 'python_yn', 'R_yn',
       'spark', 'aws', 'excel', 'Company Name', 'seniority']]
df_cat
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(18, 6), dpi=80, facecolor='w', edgecolor='k')
for i in df_cat.columns:
    cat_num = df_cat[i].value_counts()
    print("graph for %s: total = %d" % (i, len(cat_num)))
    figure(num=None, figsize=(18, 6), dpi=80, facecolor='w', edgecolor='k')
    chart = sns.barplot(x=cat_num.index, y=cat_num)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=-70)
    plt.grid()
    plt.show()
```
